```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
article_file = "Resources/ArticlesApril2018.csv"

articles_df = pd.read_csv(article_file)
articles_df.head()
```

```python
comment_file = "Resources/CommentsApril2018.csv"

comment_df = pd.read_csv(comment_file)
comment_df.head()
```

```python
new_comments_data_df = comment_df[['articleID', 'commentBody', 'commentType','status', 'trusted', 'userDisplayName', 'userLocation']].copy()
new_comments_data_df.head()
```

```python
new_article_data_df = articles_df[['articleID', 'byline', 'articleWordCount','headline', 'keywords', 'typeOfMaterial']].copy()
new_article_data_df.head()
```

```python
merge_table = pd.merge(new_comments_data_df, new_article_data_df, on="articleID")
merge_table
```

```python
nyt_articles_df =merge_table.rename(index=str, columns={"commentBody": "Comment", "userDisplayName": "Username", "byline": "Author", 
                                       "articleID": "ArticleID",
                                       "commentType": "CommentType", "status": "Status", "trusted": "Trusted",
                                      "userLocation": "UserLocation", "articleWordCount": "ArticleWordCount",
                                      "headline": "Headline", "keywords": "Keywords", "typeOfMaterial":"TypeofMaterial"})

nyt_articles_df.head()
```

```python
connection_string = f"root:Muskies95!!@localhost/NYT_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
engine.table_names()
```

```python
nyt_articles_df.to_sql(name="article_comments", con=engine, if_exists="append", index=True)
```
